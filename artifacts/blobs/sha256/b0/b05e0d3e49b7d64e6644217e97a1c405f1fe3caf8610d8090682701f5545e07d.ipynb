# Discovery-chain export analysis demo

This notebook demonstrates how to use `scripts/export_and_analyze.py` to convert a PsyNet export into the requested subject, action, event, and notebook CSV files.

```python
from pathlib import Path
import subprocess

experiment_dir = Path("../../code/discovery_game").resolve()
export_zip = Path("../simulated_data.zip").resolve()
output_dir = Path("../derived_csv").resolve()
cmd = [
    "python",
    str(experiment_dir / "scripts" / "export_and_analyze.py"),
    "--export-path", str(export_zip),
    "--output-dir", str(output_dir),
]
print("Running:", " ".join(cmd))
subprocess.run(cmd, check=True)
print("Output files:", sorted(p.name for p in output_dir.glob("*.csv")))
```

```python
import csv
from pathlib import Path

def read_rows(name, n=3):
    path = output_dir / name
    with path.open(newline="", encoding="utf-8") as f:
        rows = list(csv.DictReader(f))
    print(f"{name}: {len(rows)} rows")
    for row in rows[:n]:
        print(row)
    print()

for filename in ["subject_data.csv", "action_data.csv", "event_data.csv", "notebook_data.csv"]:
    read_rows(filename)
```
