# Cross-cultural decision-making simulation analysis

This notebook summarizes the simulated PsyNet export generated by `psynet simulate`.

```python
from pathlib import Path
import pandas as pd
root = Path('..').resolve()
trial_csv = root / '..' / 'code' / 'decision_making_study' / 'data' / 'simulated_data' / 'regular' / 'basic_data' / 'trial.csv'
participant_csv = root / '..' / 'code' / 'decision_making_study' / 'data' / 'simulated_data' / 'regular' / 'basic_data' / 'participant.csv'
trials = pd.read_csv(trial_csv)
participants = pd.read_csv(participant_csv)
summary = {
    'participants': participants['participant_id'].nunique(),
    'trial_rows': len(trials),
    'unique_trials': trials['trial_id'].nunique(),
    'choices': sorted(trials['choice'].dropna().unique()),
    'all_trials_complete': trials['complete'].all(),
    'all_trials_finalized': trials['finalized'].all(),
}
summary
```
